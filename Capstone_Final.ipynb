```python
asheville_latitude, asheville_longitude
```

```python
# search foursquare for asheville venues

# setting up LIMIT and radius foursquare search parameters
LIMIT = 500 
radius = 3200

# create URL 
url1 = 'https://api.foursquare.com/v2/venues/explore?&client_id=E3ZQ5HY2IJMKKLIRD4VO4CIMALQJSMIN3KPEHQQYS4Z3FU22&client_secret=TXA2E5OSXNLRHITNY4FTUJVWG0OQZR0F1ZTFRN1TZ1L1DNZQ&v=20180605&ll=35.6009498,-82.5540161&radius=3200&limit=500'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    asheville_latitude, 
    asheville_longitude, 
    radius, 
    LIMIT)

# display URL
url1 
```

```python
# get the results of the search and store in variable
asheville_results = requests.get(url1).json()
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# setting up venues group
venues = asheville_results['response']['groups'][0]['items']
    
asheville_nearby_venues = pd.json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
asheville_nearby_venues = asheville_nearby_venues.loc[:, filtered_columns]

# filter the category for each row
asheville_nearby_venues['venue.categories'] = asheville_nearby_venues.apply(get_category_type, axis=1)

# clean columns
asheville_nearby_venues.columns = [col.split(".")[-1] for col in asheville_nearby_venues.columns]

# display part of the DataFrame
asheville_nearby_venues.head(20)
```

```python
# get shape of asheville DataFrame
asheville_nearby_venues.shape
```

```python
print('{} Asheville venues were returned by Foursquare.'.format(asheville_nearby_venues.shape[0]))
```

```python
# count categories in DataFrame
asheville_counts = asheville_nearby_venues.groupby('categories').count()

# drop 'lat' and 'lng' columns
asheville_counts.drop(['lat','lng'], axis=1, inplace=True)

# print revised asheville DataFrame
asheville_counts
```
